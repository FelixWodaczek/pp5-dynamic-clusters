```python
import pandas as pd
import numpy as np
import json

from pathlib import Path
```

```python
from pp5.prec import ProteinRecord
from pp5.align import Arpeggio
from pp5.external_dbs.pdb import split_id

interaction_cutoff = 4.5
# adjacency_cutoff = 3
pdb_id = "2WUR:A"

arpeggio = Arpeggio(interaction_cutoff=interaction_cutoff, use_conda_env="arpeggio", cache=True)
df_arp = arpeggio.contact_df(pdb_id)
```

```python
df_arp.to_csv(f'out/arpeggio/arp-{pdb_id}-{interaction_cutoff}.csv')
df_arp
```

```python
prec = ProteinRecord.from_pdb(pdb_id)
```

```python
# Create a temp df to work with
df = df_arp.copy().reset_index()

# Convert 'contact' column to text
df['contact'] = df['contact'].apply(lambda x: str.join(',', sorted(x)))
df
```

```python
# Ignore any water contacts
idx_non_water = ~df["interacting_entities"].isin(["SELECTION_WATER", "NON_SELECTION_WATER", "WATER_WATER"])
sum(idx_non_water)/len(idx_non_water)
```

```python
# Ignore contacts which are of type 'proximal' only
idx_non_proximal_only = df['contact'] != "proximal"
sum(idx_non_proximal_only)/len(idx_non_proximal_only)
```

```python
# Ignore contacts starting from another chain
pdb_base_id, pdb_chain_id = split_id(pdb_id)
idx_non_other_chain = df['bgn.auth_asym_id'].str.lower() == pdb_chain_id.lower()

sum(idx_non_other_chain)/len(idx_non_other_chain)
```

```python
# Find contacts ending on other chain
idx_end_other_chain = df['end.auth_asym_id'].str.lower() != pdb_chain_id.lower()
print(sum(idx_end_other_chain)/len(idx_end_other_chain))

contact_any_ooc = df['end.auth_asym_id'].copy()
contact_any_ooc[~idx_end_other_chain] = ''
```

```python
# Calculate sequence distance for each contact
contact_sequence_dist = (df['end.auth_seq_id'] - df['bgn.auth_seq_id']).abs()

# If end is not on chain, set sdist to NaN to clarify that it's invalid
contact_sequence_dist[idx_end_other_chain] = float("nan")
contact_sequence_dist
```

```python
from pp5.codons import ACIDS_3TO1, ACIDS_1TO3

# Find interactions with non-AAs (ligands)
contact_non_aa = df['end.label_comp_id'].copy()
idx_end_non_aa = ~contact_non_aa.isin(list(ACIDS_3TO1.keys()))
contact_non_aa[~idx_end_non_aa] = ''

idx_end_non_aa
print(sum(idx_end_non_aa)/len(idx_end_non_aa))
contact_non_aa
```

```python
# Filter only contacting
df_filt = df[idx_non_water & idx_non_proximal_only & idx_non_other_chain]

# Note: this assign works because after filtering, the index remains intact
df_filt = df_filt.assign(
    contact_sdist=contact_sequence_dist,
    contact_any_ooc=contact_any_ooc,
    contact_non_aa=contact_non_aa
)

df_filt = df_filt.drop('bgn.auth_asym_id', axis='columns')
df_filt = df_filt.astype({"bgn.auth_seq_id": str})
df_filt = df_filt.set_index(["bgn.auth_seq_id"])
df_filt = df_filt.sort_values(by=["bgn.auth_seq_id"])
df_filt = df_filt.rename_axis('res_id')
df_filt.to_csv(f'out/arpeggio/arp-filt-{pdb_id}-i{interaction_cutoff}.csv')
df_filt
```

```python

df_prec = prec.to_dataframe()
    
df_prec_join_filt = df_prec.join(
# df_prec[['res_id','name']].join(
    #df_filt.reset_index().astype({'bgn.auth_seq_id': str}).set_index('bgn.auth_seq_id').drop(labels=[
    # df_filt.drop(labels=[
    #     'end.auth_asym_id',
    #     'end.auth_seq_id',
    #     'end.auth_atom_id',
    #     'end.label_comp_id',
    #     'end.pdbx_PDB_ins_code',
    #     'interacting_entities',
    #     'bgn.auth_atom_id',
    # ], axis='columns'),
    df_filt,
    how='left',
    on='res_id',
).set_index('res_id')

df_prec_join_filt['bgn.label_comp_id'] = df_prec_join_filt['bgn.label_comp_id'].apply(lambda aaa: ACIDS_3TO1.get(aaa, 'X'))

idx_aa_mismatch = df_prec_join_filt['name'] != df_prec_join_filt['bgn.label_comp_id']

# Make sure that joining produces corresponding residues
# for i, (res_id, res_id_) in enumerate(zip(df_prec_join_filt['name'],df_prec_join_filt['bgn.label_comp_id'])):
#     assert res_id == res_id_ or pd.isna(res_id_), f'#{i}: {res_id=}, {res_id_=}'
    
df_mismatch = df_prec_join_filt[idx_aa_mismatch]
assert all(pd.isna(df_mismatch['contact']))
df_mismatch
```

```python
from itertools import chain

# Aggregate contacts per AA

def _agg_contact(items):
    return str.join(',', sorted(set(chain(*[str.split(it, ',') for it in items if it]))))

df_groups = df_filt.groupby(by=['res_id']).aggregate({
    'contact': ['count', _agg_contact],
    'distance': ['min','max',],
    # min and max will ignore nans, and the lambda will count them
    'contact_sdist': ['min','max'],
    'contact_any_ooc': [_agg_contact],
    'contact_non_aa': [_agg_contact],
}
)

df_contacts = df_groups.set_axis(
    labels=[
        'contact_count','contact_types',
        'contact_dmin', 'contact_dmax',
        'contact_smin', 'contact_smax',
        'contact_ooc',
        'contact_non_aa',
    ], axis='columns')
df_contacts = df_contacts.reset_index().astype({'res_id': str}).set_index('res_id')
df_contacts
```

```python
df_contacts[(df_contacts['contact_ooc']==True) | (~df_contacts['contact_non_aa'].isna())]
```

```python
df_prec_contacts = prec.to_dataframe().join(
    df_contacts,
    how='left',
    on='res_id',
)

df_prec_contacts['contact_count'].fillna(0, inplace=True)
df_prec_contacts = df_prec_contacts.astype({'contact_count': int, 'contact_smin': pd.Int64Dtype(), 'contact_smax': pd.Int64Dtype()})
df_prec_contacts = df_prec_contacts.set_index('res_id')

df_prec_contacts.to_csv(f'out/arpeggio/prec-arp-{pdb_id}-i{interaction_cutoff}.csv')

df_prec_contacts
```
